```python
def exchange_rates(currency, format='full'):    
    url = 'https://www.cbr-xml-daily.ru/daily_json.js'  
    response = requests.get(url).json()['Valute']    
    data = response[currency]     
    if format == 'full':    
        return data      
    elif format == 'value':    
        return data['Value']
```

```python
exchange_rates( 'EUR', format = 'value' )
```

```python
def currency_name(ID):    
    url = 'https://www.cbr-xml-daily.ru/daily_json.js'  
    response = requests.get(url).json()['Valute']    
    for data in response.values():
        if data['ID'] == ID:
            return data['Name']
```

```python
from bs4 import BeautifulSoup
import requests
def get_actors(url):
    response = requests.get(url)
    page = BeautifulSoup(response.text, 'html.parser')
    acts = page.find('div', id="actorList")
    act = acts.find_all('a')
    list_act = []
    for acter in act[:]:
        list_act.append(acter.text)
    return (list_act)
get_actors('https://www.kinopoisk.ru/film/42326/')
```

```python
def get_actors(url):

    response = requests.get(url)

    page = BeautifulSoup(response.text, 'html.parser')

    art_list = page.find('div', id = 'actorList')

    artists = art_list.find_all('a')

    return [artist.text for artist in artists]

print(get_actors('https://www.kinopoisk.ru/film/448/'))
```

```python
def get_smm_index(group_name, token):
    stats = {}  
    comments = 0
    likes = 0
    reposts = 0
    views = 0
    url = 'https://api.vk.com/method/wall.get'  
    params = {  
        'domain': group_name,  
        'filter': 'owner',  
        'count': 10,  
        'offset': 0,  
        'access_token': token,  
        'v': 5.95  
    }  
    response = requests.get(url, params = params)
    for record in response.json()['response']['items'][-10:-1]:  
          
        comments += record['comments']['count'] 
        likes += record['likes']['count'] 
        reposts += record['reposts']['count']
        views += record['views']['count']
    smm_index = (comments + likes + reposts) / views
    return smm_index 
```
